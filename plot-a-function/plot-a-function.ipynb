# Plot a function in Julia, basics

```julia
# there are various backends,
# for interactive plots in notebooks use Plotly
# (cannot export Plotly plots to html...)
using Plots
pyplot()
Plots.PyPlotBackend();
```

To plot a function $f: \mathbb{R}^2 \to \mathbb{R}$ as a set of points in $\mathbb{R}^3$, import the package `Plots` and use the function `surface`

```julia
f(x,y) = x^2 - y^2
```

```julia
surface(-10:2:10, -10:2:10, f)
```

where $f$ can also be an anonymous function.

```julia
surface(-50:5:50, -10:0.5:10, (x,y) -> x^2 +  y^3)
```

To plot a heatmap use `heatmap`, pass arguments the same way as with `surface`.

```julia
heatmap(1:10,1:20,(x,y) -> 9x + 6y)
```

To plot a contour plot use `contour`.

```julia
contour(-10:10, -10:10, f)
```

For example, you can plot a function $\gamma: \mathbb{R} \to \mathbb{R}^2$ like this.

```julia
γ(t) = (2 * cos(t), sin(t))
```

```julia
values = [γ(x) for x in -π:0.1:π]
# scatter plot
scatter(values, aspect_ratio = 1)
```

```julia
# subsequent points connected by a straight line
plot(values, aspect_ratio = 1)
```

Further, a function $f: \mathbb{R} \to \mathbb{R}^3$, can be plotted like this.

```julia
g(t) = (sin(t), cos(t), 0.2t)
```

```julia
values = [g(t) for t in 0:0.1:2π]
plot(values)
```
